# Duplicated data

Now that we have deleted incorrect data, we have to go through our data to check for duplicates. We do this by looping through the data and deleting each duplicate app. For example the following data is duplicated:

```python
duplicated = []
apps = []

for app in android_data:
    if app[0] in apps:
        duplicated.append(app[0])
    else:
        apps.append(app[0])

#This one cleans the duplicated list, so that each duplicate is shown only once
clean_duplicated = []
for app in duplicated:
    if not app in clean_duplicated:
        clean_duplicated.append(app)
        
print("Some apps that are duplicated in the Google Play dataset:")
print(clean_duplicated[0:10])
```

As we can see, there are multiple duplicated apps in the data set. The total amount of duplicates can be seen with the `len()` function.

```python
duplicates = str(len(duplicated))
print("There are a total of " + duplicates + " duplicates in the data set.")
```

We need to remove these duplicates, but we don't want to do that randomly. There are duplicates because the data has been taken at different times, so the fourth column (amount of ratings) has differences. We will keep the one with the highest amount of ratings, because more ratings means more accuracy. To do that, we need to create a dictionary, and loop through the data, adding only the highest rating data to the dictionary.

```python
reviews_max = {}

for app in android_data:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif not name in reviews_max:
        reviews_max[name] = n_reviews
```

We can check if the code works by checking the length of the dictionary. If we deduct the duplicates from the length of the original data, we can see what the expected value should be:

```python
expected = len(android_data) - int(duplicates)
print("Expected value: " + str(expected))
print("The length of our dictionary: " + str(len(reviews_max)))
```

Now we have to remove the duplicate rows. We do this by creating two lists.

```python
android_clean = []
already_added = []

for app in android_data:
    name = app[0]
    n_reviews = float(app[3])
    
    if n_reviews == reviews_max[name] and name not in already_added:
        android_clean.append(app)
        already_added.append(name)
```

Now we can try if the data is correct by checking the length, which should be 9659. Let's also print a few rows to see that everything is correct.

```python
print("The length of clean data is: " + str(len(android_clean)))
print("Let's test 3 rows...")
print(android_clean[0:2])
```

The App store data doesn't have any duplicates, so we can move on to deleting apps that are not designed for English-speaking audience. This can be done by using ASCII, where english text ranges between 0 and 127. For example when we type `ord("a")`, we get 97, because that is the numerical value for a.

```python
def language_example(word): 
    for character in word:
        if ord(character) > 127:
            return False
    return True
    
```

```python
print(language_example("Instagram"))
```

However, this method has a problem. A lot of apps include things like emojis in their names, which have ASCII value over 127. This deletes a lot of useful data, for example:

```python
print(language_example("Instachat üòú"))
```

Because of this, we are only going to delete data that has more than three characters falling outside the 0-127 ASCII range.

```python
def english(word):
    amount = 0
    
    for character in word:
        if ord(character) > 127:
            amount += 1
            
    if amount > 3:
        return False
    else:
        return True
```

```python
print(english('Docs To Go‚Ñ¢ Free Office Suite'))
print(english('Instachat üòú'))
print(english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
```

Now that we have a working function, we can use it to remove rows from our datasets. We loop through both of the datasets, App Store and Google Play.

```python
#Android
english_apps_android = []
for app in android_clean:
    word = app[0]
    if english(word) == True:
        english_apps_android.append(app)
        
#Apple
english_apps_apple = []
for app in apple_data:
    word = app[1]
    if english(word) == True:
        english_apps_apple.append(app)
        
explore_data(english_apps_android, 0, 3, True)
print("\n")
explore_data(english_apps_apple, 0, 3, True)
```

The last part of data cleaning is to remove the apps that are not free. This can be done by looping throught each dataset, and removing those that have a higher price than 0. The prices are as a string, so we are using if statements to determine whether the price is equal to "0" or "0.0".

```python
#Android
android_final = []
for app in english_apps_android:
    price = app[7]
    if price == "0":
        android_final.append(app)
        
#Apple
apple_final = []
for app in english_apps_apple:
    price = app[4]
    if price == "0.0":
        apple_final.append(app)

print(len(android_final))        
print(len(apple_final))
```

We are left with 8864 Android apps and 3222 Apple apps.

# Most Common Apps

Our end goal is to create an app for both iOS and android. The strategy is following:
1. Build a minimal Android version of the app, and add it to Google Play.
2. If the app has a good response from users, we develop it further.
3. If the app is profitable after six months, we build an iOS version of the app and add it to the App Store.

Because of this we have to find out which genres do well in Google Play and in App Store.

To determine this, we are going to use columns `Genres` in Google Play Store data, and `prime_genre` in Apple Store data.

```python
def freq_table(dataset, index):
    frequency_table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in frequency_table:
            frequency_table[value] += 1
        else:
            frequency_table[value] = 1
            
    table_percentages = {}
    for key in frequency_table:
        frequency_table[key] = (frequency_table[key]/total) * 100
    
    return frequency_table
```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

To find out which of the categories/genres has the most downloads, we can use `Installs` column in Google Play Store data. This isn't available in Apple Store data, so we are going to use the number of ratings each app has. To do that, we have to loop through each genre.

```python
#Let's generate a frequency table for prime_genre column.
prime_genres = freq_table(apple_final, 11)
for genre in prime_genres:
    total = 0
    len_genre = 0
    for app in apple_final:
        genre_app = app[11]
        if genre_app == genre:
            n_ratings = float(app[5])
            total += n_ratings
            len_genre += 1
    avg_ratings = total / len_genre
    print(genre, ":", avg_ratings)       
```

According to this data, the highest reviews are for `Navigation`. After that we have `Reference` and `Social Networking`. We should notice that most of these categories are heavily influenced by apps like Waze, Google Maps and Geocaching(Navigation), or Facebook, Pinterest, Skype (Social networking). This makes the categories seem more popular than they are in reality.

With Google Play Store data we have a lot of open-ended data (100+, 1,000+ etc.). Because of this we have to get rid of the commas and plus signs and convert the data from string to float.

```python
#Let's generate a frequency table for Category column.
android_categories = freq_table(android_final, 1)
for category in android_categories:
    total = 0
    len_category = 0
    for app in android_final:
        category_app = app[1]
        if category == category_app:
            n_installs = app[5]
            n_installs = n_installs.replace(",", "")
            n_installs = n_installs.replace("+", "")
            n_installs = float(n_installs)
            
            total += n_installs
            len_category += 1
            
    average_installs = total/len_category
    print(category, ":", average_installs)
```

On average, communication apps have the most installs: 38,456,119. This is heavily skewed by apps like WhatsApp, Skype, Facebook Messenger etc. This can be seen by removing all the apps that have over 100 million installs:

```python
under_100_m = []

for app in android_final:
    n_installs = app[5]
    n_installs = n_installs.replace(',', '')
    n_installs = n_installs.replace('+', '')
    if (app[1] == 'COMMUNICATION') and (float(n_installs) < 100000000):
        under_100_m.append(float(n_installs))
        
sum(under_100_m) / len(under_100_m)
```
